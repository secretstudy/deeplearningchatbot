# 직선그래프

```python
import matplotlib.pyplot as plt

x = [a for a in range(0,11)]
y = list(range(0,11))

print('x축',x)
print('y축',y)

plt.plot(x, y)
plt.show()
```

# 2차 함수 그래프

```python
import matplotlib.pyplot as plt

f = lambda x: x ** 2

x = [x for x in range(-8, 9)]
y = [f(y) for y in range(-8, 9)]

print('x축',x)
print('y축',y)

plt.plot(x,y)
plt.show()
```

# 계절별 온도 비차트

```python
import matplotlib.pyplot as plt

temperatures = [3.3,34.5,14.2,-10]
x = list(range(4))
x_labels = ['Spring','Summer','Fall','Winter']

plt.title("Bar Chart")
plt.bar(x, temperatures)
plt.xticks(x, x_labels)
plt.yticks(sorted(temperatures))
plt.xlabel("season")
plt.ylabel("temperature")
plt.show(temperatures)
```
